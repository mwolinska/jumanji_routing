```python
from ic_routing_board_generation.ic_rl_training.online_generators.random_seed_generator import RandomSeedGenerator
import jax
```

```python
random_seed_generator = RandomSeedGenerator(10, 5)
key = jax.random.PRNGKey(0)
```

```python
#Random seed generator
random_seed_call = jax.jit(random_seed_generator.__call__)
%time jax.block_until_ready(random_seed_call(key))  # compilation
```

```python
%timeit jax.block_until_ready(random_seed_call(key))  # using cache
```

```python
random_seed_generator = RandomSeedGenerator(10, 5)
batch_size = 100
keys = jax.random.split(jax.random.PRNGKey(0), batch_size)
```

```python
#random seed generator
random_seed_call = jax.jit(jax.vmap(random_seed_generator.__call__))
%time jax.block_until_ready(random_seed_call(keys))  # compilation
```

```python
%timeit jax.block_until_ready(random_seed_call(keys))  # using cache
```
